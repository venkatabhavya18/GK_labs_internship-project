```python
import pdfplumber
import os

# This line allows file upload from your system
from google.colab import files
uploaded = files.upload()  # Upload your PDFs here

# Keyword to search (you can change this)
keyword = input("Enter keyword to search: ").lower()

# Create a summary list
summary = []

# Scan each uploaded PDF
for filename in uploaded.keys():
    if filename.endswith(".pdf"):
        with pdfplumber.open(filename) as pdf:
            for i, page in enumerate(pdf.pages):
                text = page.extract_text()
                if text and keyword in text.lower():
                    # Get sentences containing the keyword
                    lines = text.split('\n')
                    for line in lines:
                        if keyword in line.lower():
                            summary.append(f"{filename}, Page {i+1}: {line.strip()}")

# Save summary to file
summary_filename = "keyword_summary.txt"
with open(summary_filename, "w", encoding="utf-8") as f:
    for item in summary:
        f.write(f"{item}\n")

print(f"âœ… Summary saved to {summary_filename} ({len(summary)} matches found)")

```
